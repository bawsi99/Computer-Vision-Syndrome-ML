```python
#Importing Libraries 
import numpy as np
import pandas as pd
#Importing Dataset
dataset = pd.read_csv(r"C:\Aaryan\PROJECT\dataset.csv")
```

```python
dataset
```

```python
#Assigning numbers to word input in Gender column for betterment analysis.
dataset = dataset.replace(to_replace='m', value=1)
dataset = dataset.replace(to_replace='f', value=2)
```

```python
#Assigning numbers to word input in Eye Strain column for betterment analysis.
dataset = dataset.replace(to_replace='high', value=2)
dataset = dataset.replace(to_replace='medium', value=1)
dataset = dataset.replace(to_replace='low', value=0)
```

```python
#Assigning numbers to word input in Anti-Glare Filter column for betterment analysis.
dataset = dataset.replace(to_replace='yes', value=0)
dataset = dataset.replace(to_replace='no', value=1)
dataset
```

```python
from sklearn.model_selection import train_test_split
X = dataset.drop(["cvs"], axis=1)
y = dataset["cvs"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
#Applying Logistic Regression to the dataset
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
y_test
```

```python
#Finding the accuracy of the model
print("Accuracy : {}".format(clf.score(X_test,y_test)))
```

```python
#Applying Gaussian Naive Bayes to our dataset
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
#Predicting the output
y_pred=nb.predict(X_test)
```

```python
#Finding the accuracy of the model
print("Accuracy : {}".format(nb.score(X_test,y_test)))
```
